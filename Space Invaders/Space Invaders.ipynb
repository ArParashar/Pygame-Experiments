# Space Invader

```python
import pygame 
from pygame import mixer
import random
import math

pygame.init()

#Create screen
screen=pygame.display.set_mode((1600,900))
pygame.display.set_caption("Space Invaders")
icon=pygame.image.load("ufo.png")
background=pygame.image.load("space5.jpg")
pygame.display.set_icon(icon)

#music
mixer.music.load("score.mp3")
mixer.music.play(-1)


#Player
playericon=pygame.image.load("player1.png")
playerX=800
playerY=750
playerX_change=0

#Enemy
enemyicon=pygame.image.load("ufo.png")
enemyX=random.randint(100,1500)
enemyY=random.randint(100,500)
enemyX_change=.5
enemyY_change=40

#Bullet
bulleticon=pygame.image.load("smallbullet.png")
bulletX=0
bulletY=750
bulletX_change=0
bulletY_change=5
bullet_state="ready"   #ready =invisble, fire=moving and visible

#score
score_value=0
font=pygame.font.SysFont('Impact', 50)
textX=25
textY=25

def show_score(x,y):
    score=font.render("Score :"+ str(score_value), True,(255,255,255))
    screen.blit(score,(x,y))

def player(x,y):
    screen.blit(playericon,(x,y))
    
def enemy(x,y):
    screen.blit(enemyicon,(x,y))  
    
def fire_bullet(x,y):
    global bullet_state
    bullet_state="fire"
    screen.blit(bulleticon,(x+25,y+5))
    
def headshot(enemyX,enemyY,bulletX,bulletY):
    dist=math.sqrt(math.pow(bulletX-enemyX,2)+(math.pow(bulletY-enemyY,2)))
    if dist<20:
        return True

running=True
while running:
    
    screen.fill((0,0,0))
    screen.blit(background,(-80,0))
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            running= False
         
        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_LEFT:
                playerX_change=-2
            if event.key==pygame.K_RIGHT:
                playerX_change=+2
            if event.key==pygame.K_SPACE:
                if bullet_state=="ready":
                    bullet_sound=mixer.Sound("laser.mp3")
                    bullet_sound.play()
                    bulletX=playerX
                    fire_bullet(bulletX,bulletY)
                    
                       
        if event.type==pygame.KEYUP:  
            if event.key==pygame.K_LEFT or event.key==pygame.K_RIGHT:
                playerX_change=0
                
                
    
    #Spaceship boundary
    playerX+=playerX_change
    if playerX<=100:
        playerX=100
    elif playerX>=1500:
        playerX=1500
        
    #Enemy boundary
    enemyX+=enemyX_change
    if enemyX<=100:
        enemyX_change =1
        enemyY +=enemyY_change
    elif enemyX>=1500:
        enemyX_change -=1
        enemyY +=enemyY_change
    
    #Bullet
    if bulletY<=0:
        bulletY=750
        bullet_state="ready"
    
    if bullet_state == "fire":
        fire_bullet(bulletX,bulletY)
        bulletY-=bulletY_change
        
    #collision impact 
    collision=headshot(enemyX,enemyY,bulletX,bulletY)
    if collision:
        bulletY=750
        bullet_state="ready"
        score_value+=1
        enemyX=random.randint(100,1500)
        enemyY=random.randint(100,500)
        explosion_sound=mixer.Sound("explode.wav")
        explosion_sound.play()
    
    player(playerX,playerY)
    enemy(enemyX,enemyY)
    show_score(textX,textY)
    pygame.display.update()
```
